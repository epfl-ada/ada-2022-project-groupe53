```python
with open('data/df_paths_finished.pickle', 'rb') as handle:
    df_paths_finished = pickle.load(handle)
```

```python
df_paths_finished
```

```python
mate_in_one = pd.DataFrame(columns=['article', 'target', 'article_similarity', 'title_similarity', 'link_position', 'nb_links' ])
found_mate = pd.DataFrame(columns=['article', 'target', 'article_similarity', 'title_similarity', 'link_position', 'nb_links'])
counter_found = 0
counter_not_found = 0

for i in range(len(df_paths_finished)):
    end =  df_paths_finished.iloc[i,2]

    path = df_paths_finished.iloc[i,0]
    path_sim_article = df_paths_finished.iloc[i,4]
    path_sim_title = df_paths_finished.iloc[i,5]

    for j in range(len(path)-1): 
        # check if truth graph.article matrix has entry path[j], end 
        if end in truth_graph.matrix_articles[path[j]].keys():
            
            position = links_df.loc[path[j], end]['position']
            nb_links = links_df.loc[path[j], end]['nb_links']
            if j == len(path)-2 :
                found_mate.loc[counter_found] = [path[j], end, path_sim_article[j], path_sim_title[j], position, nb_links ]
                counter_found = counter_found +1 
            else :
                # iterate over all positions with index
                for i, index in enumerate(position):
                    mate_in_one.loc[counter_not_found] = [path[j], end, path_sim_article[j], path_sim_title[j], index, nb_links]
                    counter_not_found = counter_not_found+1
```

```python

print("Number of articles in mate_in_one: ", len(mate_in_one))
print()
print("Average similarity of article in mate_in_one: ", np.mean(mate_in_one['article_similarity']))
print("Median similarity of article in mate_in_one: ", np.median(mate_in_one['article_similarity']))
print()
print("Average similarity of title in mate_in_one: ", np.mean(mate_in_one['title_similarity']))
print("Median similarity of title in mate_in_one: ", np.median(mate_in_one['title_similarity']))
print()
print("Average position of link in mate_in_one: ", np.mean(mate_in_one['link_position']))
print("Median position of link in mate_in_one: ", np.median(mate_in_one['link_position']))
print()
print("Average number of links in mate_in_one: ", np.mean(mate_in_one['nb_links']))
print("Median number of link in mate_in_one: ", np.median(mate_in_one['nb_links']))
mate_in_one.sample(10)
```

```python
#keep only rows where link_position is of length 1
found_mate_filter = found_mate[found_mate['link_position'].apply(lambda x: len(x) == 1)]
# convert link_position to int
found_mate_filter['link_position'] = found_mate_filter['link_position'].apply(lambda x: x[0]) 
```

```python

print("Number of articles in found_mate_filter: ", len(found_mate_filter))
print()
print("Average similarity of article in found_mate_filter: ", np.mean(found_mate_filter['article_similarity']))
print("Median similarity of article in found_mate_filter: ", np.median(found_mate_filter['article_similarity']))
print()
print("Average similarity of title in found_mate_filter: ", np.mean(found_mate_filter['title_similarity']))
print("Median similarity of title in found_mate_filter: ", np.median(found_mate_filter['title_similarity']))
print()
print("Average position of link in found_mate_filter: ", np.mean(found_mate_filter['link_position']))
print("Median position of link in found_mate_filter: ", np.median(found_mate_filter['link_position']))
print()
print("Average number of links in found_mate_filter: ", np.mean(found_mate_filter['nb_links']))
print("Median number of link in found_mate_filter: ", np.median(found_mate_filter['nb_links']))
found_mate_filter.sample(10)
```

```python
found_mate_filter['label'] = 1
mate_in_one['label'] = 0

# sample len(mate_in_one) rows from found_mate
found_mate_sample = found_mate_filter.sample(len(mate_in_one))

#concatenate found_mate_sample and mate_in_one
df = pd.concat([found_mate_sample, mate_in_one])
```

```python
# import logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
# do regression to classify label based on article_similarity, title_similarity
# split into train and test
# normalize data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df[['article_similarity', 'title_similarity', 'link_position', 'nb_links']])
df[['article_similarity', 'title_similarity', 'link_position', 'nb_links']] = scaler.transform(df[['article_similarity', 'title_similarity', 'link_position', 'nb_links']])
X_train, X_test, y_train, y_test = train_test_split(df[['article_similarity', 'title_similarity', 'link_position', 'nb_links']], df['label'], test_size=0.2, random_state=42)

# train model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# predict
y_pred = logreg.predict(X_test)

# import metrics
from sklearn import metrics

# print accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# show cooefficients
print("Coefficients: ", logreg.coef_)
print("Intercept: ", logreg.intercept_)


```
